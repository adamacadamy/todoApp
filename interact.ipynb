````python
from pathlib import Path  # Import Path from pathlib for working with file and directory paths

# Define the project structure as a nested dictionary
structure = {
    ".gitignore": "",  # Create an empty .gitignore file
    "app": {
        "__init__.py": "",  # Initialize the app package
        ".env": "",  # Environment variables file
        "models": {  # Models directory
            "__init__.py": "",
            "user.py": "",
            "todo.py": "",
        },
        "routes": {  # Routes directory
            "__init__.py": "",
            "auth.py": "",
            "todo.py": "",
            "views.py": ""
        },
        "schemas": {  # Schemas directory
            "__init__.py": "",
            "user_schema.py": "",
            "todo_schema.py": "",
        },
        "forms": {  # Forms directory
            "__init__.py": "",
            "login_form.py": "",
            "register_form.py": "",
            "todo_form.py": "",
        },
        "templates": {  # Templates directory (HTML files)
            "base.html": "",
            "login.html": "",
            "register.html": "",
            "todo.html": "",
        },
    },
    "static": {  # Static assets directory (CSS, JS)
        "style.css": "",
        "script.js": ""
    },
    "run.py": "",  # Main script to run the Flask app
    "requirements.txt": "",  # Dependencies file
    "migrations": {}  # Empty migrations directory
}

def create_flask_scaffold(parent: Path, structure: dict) -> None:
    """
    Recursively creates a Flask project scaffold based on the given directory structure.

    Args:
        parent (Path): The root directory where the scaffold should be created.
        structure (dict): The dictionary defining the folder and file structure.
    """
    for name, content in structure.items():  # Loop through the structure dictionary
        path = parent / name  # Construct the full path for each file/folder

        display(f"name: {name}")
        display(f"content: {content}")
        display(f"creating artifact {path}")
        display("---------------------------------------")
        if isinstance(content, dict):  # Check if the item is a directory (nested dictionary)
            path.mkdir(parents=True, exist_ok=True)  # Create the directory if it doesn't exist
            create_flask_scaffold(path, content)  # Recursively call function for subdirectories
        else:  # If it's a file (empty string value)
            path.touch(exist_ok=True)  # Create an empty file if it doesn't exist

# Call the function to generate the project structure in the current working directory
create_flask_scaffold(parent=Path.cwd()/"tt", structure=structure)
````
